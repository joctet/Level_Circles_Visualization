**Imports**

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from math import *
```

Let be a function f from ℝ² into ℝ. For any couple (x, y) in ℝ² :

```python
def f(x, y):
    return 2*x**2 - x*y + 2*y**2
```

We choose a range of values from the set of real numbers ℝ for x and y (I will take for the example [-100, 100] with 100 values for each variable).

```python
x = np.linspace(-100, 100, 100)
y = np.linspace(-100, 100, 100)
```

We will then create an array of values containing all value pairs (xi, yi) for xi in x, and yi in y.

```python
X, Y = np.meshgrid(x, y)
```

Then, we compute the values of our function defined above for each pair of values (xi, yi) and store these values in a new variable Z :


```python
Z = f(X,Y)
```

## Visualizations

```python
fig = plt.figure()
contours = plt.contour(X, Y, Z)
plt.title("Level circles", fontsize=13)
plt.xlabel('x', fontsize=11)
plt.ylabel('y', fontsize=11)
plt.show()
```

Well it's not crazy! Fortunately, we can go further and choose the number of level circles ("levels" : here levels = 5 by default), define the size of the figure (figsize), display the values of the level circles (plt.clabel), choose the size of the characters displayed on the level circles (fontsize), add a color scale (colorbar), modify the color scale (cmap : I put a list of possible values for cmap) …

```python
fig = plt.figure(figsize = (10,7))
contours = plt.contour(Z, levels = 20, cmap = 'jet')
plt.clabel(contours, inline = True, fontsize = 9)
plt.title("Level circles", fontsize=13)
plt.xlabel('x', fontsize=11)
plt.ylabel('y', fontsize=11)
plt.colorbar()
plt.show()
```

It's more beautiful!

# A variation of the level circles

We can also represent the variations of our function in 2D in a continuous way.

```python
fig = plt.figure(figsize = (10,7))
plt.imshow(Z, extent = [-100,100,-100,100], origin = 'lower', cmap = 'hot', alpha = 1)
plt.title("Continuous level circles", fontsize=13)
plt.xlabel('x', fontsize=11)
plt.ylabel('y', fontsize=11)
plt.colorbar()
plt.show()
```

We can modify the intensity of the colors (by varying alpha between 0 and 1), modify the scale of the colors by modifying cmap…
Be careful to enter the values in the extent parameter which defines the boundaries of the output image: here x and y take values in [-100, 100].
